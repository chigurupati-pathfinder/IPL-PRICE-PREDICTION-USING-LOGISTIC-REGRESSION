```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import linear_model
import math
import seaborn as sn
import matplotlib.pyplot as plt

import pandas as pd

sold_price = pd.read_csv(r'C:\Users\SOUBHAGYALAKSHMI CH\Documents\ML machine learning\IPL IMB381IPL2013.csv')

sold_price.head()

sold_price.info()

X_features = sold_price.columns.tolist()  # Convert column names to list
X_features = ['AGE'] + X_features  # Append 'AGE' to the beginning of the list

sold_price['PLAYING ROLE'].unique()

pd.get_dummies(sold_price['PLAYING ROLE'])[0:5]

categorical_features = [ 'AGE', 'COUNTRY', 'PLAYING ROLE', 'CAPTAINCY EXP']


sold_price_encoded_df = pd.get_dummies(sold_price[X_features],columns=categorical_features)

sold_price_encoded_df.columns

X_features=sold_price_encoded_df.columns

X= sm.add_constant(sold_price_encoded_df)

Y = sold_price['SOLD PRICE']
train_X, test_X, train_y, test_y = train_test_split(X, Y,train_size = 0.8, random_state = 42 )

# Identify columns with string values
columns_with_strings = train_X.select_dtypes(include=['object']).columns

# Drop columns with string values
train_X_numeric = train_X.drop(columns_with_strings, axis=1)

# Fit the model using the preprocessed data
sold_price_model_1 = sm.OLS(train_y, train_X_numeric.astype(float)).fit()
print(sold_price_model_1.summary2())

X

from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif_factors(X):
    numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
    vif_factors = pd.DataFrame()
    vif = [variance_inflation_factor(X[numeric_cols].values, i) for i in range(X[numeric_cols].shape[1])]
    vif_factors['column'] = numeric_cols
    vif_factors['VIF'] = vif
    return vif_factors

vif_fact = get_vif_factors(X[X_features])
vif_fact

columns_with_large_vif = vif_fact[vif_fact.VIF > 4].column
plt.figure( figsize = (16,10) )
sn.heatmap( X[columns_with_large_vif].corr(), annot = True )
plt.title("Heatmap depicting correlation between features")

columns_removed = ['T-RUNS', 'T-WKTS', 'ODI-RUNS-S', 'ODI-SR-B', 'RUNS-S', 'HS', 'AVE', 'SR-B', 'RUNS-C', 'AVE-BL', 'ECON', 'SR-BL','SIXERS', 'ODI-WKTS', 'ODI-SR-BL', 'WKTS']
X_new_features = list( set(X_features) - set(columns_removed))
get_vif_factors( X[X_new_features] )

X_new_features=list(set(X_features)-set(columns_removed))
get_vif_factors(X[X_new_features])


# Check the column names in train_X DataFrame
print(train_X.columns)

# Ensure that X_new_features contains the correct column names
print(X_new_features)

# Adjust X_new_features to contain the correct column names present in train_X DataFrame
X_new_features = ['COUNTRY_IND', 'COUNTRY_ENG', 'CAPTAINCY EXP_1']

# Select the desired columns from train_X DataFrame
train_X = train_X[X_new_features]

# Fit the model and summarize
sold_price_model_2 = sm.OLS(train_y, train_X).fit()
print(sold_price_model_2.summary2())

significant_vars = ['COUNTRY_IND', 'COUNTRY_ENG', 'CAPTAINCY EXP_1']

train_X = train_X[significant_vars]

ipl_model_3 = sm.OLS(train_y, train_X).fit()

ipl_model_3.summary2()

train_Y=np.sqrt(train_y)
sold_price_model_4=sm.OLS(train_Y,train_X).fit()

sold_price_model_4.summary2()
```
